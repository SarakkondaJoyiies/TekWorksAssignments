***Bernoulli Distribution***

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
p=0.6
data = bernoulli.rvs(p, size=1000)
plt.bar([0,1], [sum(data == 0), sum(data == 1)], color=['red','green'], tick_label=["Failure (0)", "Success (1)"])
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.title("Bernoulli Distribution (p=0.6)")
plt.show()
```

***Normal Distribution***

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
mu=70
sigma=10
x=np.linspace(mu-4*sigma, mu+4*sigma, 100)
pdf=norm.pdf(x, mu, sigma)
plt.plot(x,pdf,color='skyblue',label=f"N({mu}, {sigma})")
plt.xlabel("Exam Scores")
plt.ylabel("Probability Density")
plt.title(f"Normal Distribution (={mu}, ={sigma})")
plt.legend()
plt.grid()
plt.show()
```

***Standard Normal Distribution***

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
z=np.linspace(-4, 4, 1000)
pdf=norm.pdf(z, 0, 1)
plt.plot(z, pdf, color='blue', label="Standard Normal Distribution (=0,=1)")
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.title("Standard Normal Distribution(Z-Distribution)")
plt.axvline(x=0, color='red', linestyle='--', label="Mean (Z=0)")
plt.legend()
plt.grid()
plt.show()
```

***Central Limit Theorem***

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(42)
income_data=np.random.exponential(scale=50, size=1000)
plt.figure(figsize=(8,4))
sns.histplot(income_data, bins=50, kde=True, color="g")
plt.title("Original Income Distribution (Skewed)")
plt.xlabel("Income ($1000)")
plt.ylabel("Frequency")
plt.show()

sample_means=[]
n_samples=1000
sample_size=30

for _ in range(n_samples):
    sample=np.random.choice(income_data, size=sample_size)
    sample_means.append(np.mean(sample))
plt.figure(figsize=(8, 4))
sns.histplot(sample_means, bins=50, kde=True, color="b")
plt.title(f"Sampling Distribution of Sample Mean (n={sample_size})")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```
