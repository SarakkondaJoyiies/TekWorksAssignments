```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[5,6,7,8,9]
plt.plot(x,y)
plt.title("Simple line plot")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[5,6,7,8,9]
plt.plot(x,y,linestyle='dotted',color='magenta',linewidth=3,marker='*')
plt.title("Simple line plot")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[0,10000,20000,30000,50000,60000]
plt.plot(x,y,linestyle='dashed',color='green',linewidth=2,marker='*')
plt.title("Growth of a savings account")
plt.xlabel('Months(1-6)')
plt.ylabel('Savings amount')
plt.show()
```

```python
x=['A','B','C','D','E']
y=[10,15,7,12,9]
plt.bar(x,y,color='darkred')
plt.title("Basic Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
x=['A','B','C','D','E']
y=[10,15,7,12,9]
plt.bar(x,y,color=['blue','orange','forestgreen','purple','yellow'])
plt.title("Basic Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
x=['A','B','C','D','E']
y=[10,15,7,12,9]
plt.bar(x,y,color=['skyblue','lightsalmon','lightgreen','thistle','lightyellow'],edgecolor='black')
plt.title("Basic Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
ages = [20,22,22,23,25,27,27,28,30,31,32,34,35,35,35]
b = [20,30,40,50,60,70]
plt.hist(ages,bins=b,color='mistyrose',edgecolor='black')
plt.title("Age Distribution of Customers")
plt.xlabel("Age Ranges")
plt.ylabel("Frequency")
plt.show()
```

```python
lang_names = ['Python','Ruby','C++','Java']
percentage = [40,10,20,30]
plt.pie(percentage,labels=lang_names,autopct='%1.1f%%',colors=['olive','rosybrown','slateblue','mediumvioletred'],startangle=90)
plt.title("Programming Language Usage")
plt.legend()
plt.show()
```

```python
x=[5,10,15,20,25]
y=[10,20,25,30,35]
plt.scatter(x,y,color='gold',s=100,edgecolor='black')
plt.title("Scatter Plot Example")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```
