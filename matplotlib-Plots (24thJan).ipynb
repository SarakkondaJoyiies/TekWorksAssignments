```python
#Plotting graphs using "Sales Data Analysis" dataset(from kaggle)
```

```python
import pandas as pd
data=pd.read_csv("C://Users//JOYIIES//OneDrive//Desktop//COE training//Sales Data.csv")
print(data.dtypes)
data.columns
```

```python
#Histograms for order quantity distribution

plt.figure(figsize=(10, 6))
plt.hist(data['Quantity Ordered'], bins=30, color='peachpuff', edgecolor='black')
plt.title('Distribution of Order Quantities')
plt.xlabel('Quantity Ordered')
plt.ylabel('Frequency')
plt.show()
```

```python
#Line Plots to track monthly sales

data.rename(columns={'Order_date': 'Order Date'}, inplace=True)
data['Order Date'] = pd.to_datetime(data['Order Date'], errors='coerce')

data['Quantity Ordered'] = pd.to_numeric(data['Quantity Ordered'], errors='coerce')
data['Price Each'] = pd.to_numeric(data['Price Each'], errors='coerce')

data['Year'] = data['Order Date'].dt.year
data['Month'] = data['Order Date'].dt.month

data['Sales'] = data['Quantity Ordered'] * data['Price Each']
monthly_sales = data.groupby(['Year', 'Month'])['Sales'].sum().reset_index()

plt.figure(figsize=(10, 6))
for year in monthly_sales['Year'].unique():
    data = monthly_sales[monthly_sales['Year'] == year]
    plt.plot(data['Month'], data['Sales'], marker='o')
plt.title('Monthly Sales Over Time')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

```python
##Plotting graphs using "School_performance" dataset (created on my own)
```

```python
data = pd.read_csv("C://Users//JOYIIES//OneDrive//Desktop//COE training//School_performance.csv")
```

```python
# Bar chart comparing marks across subjects for each student

subjects = ['Math', 'English', 'Spanish', 'History', 'Geography', 'Physics', 'Chemistry', 'Biology', 'Computer_Science']
plt.figure(figsize=(10, 5))
data.set_index('Name')[subjects].plot(kind='bar', figsize=(12, 8))
plt.title('Comparison of Student Marks Across Subjects')
plt.xlabel('Student Names')
plt.ylabel('Marks')
plt.xticks(rotation=90)
plt.legend(title="Subjects")
plt.tight_layout()
plt.show()
```

```python
# Box plot to identify outliers across subjects

import seaborn as sns
plt.figure(figsize=(12, 6))
sns.boxplot(data=data[subjects])
plt.title('Outliers in Student Marks Across Subjects')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.tight_layout()
plt.show()
```
