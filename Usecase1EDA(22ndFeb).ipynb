```python
# Function to detect outliers using the Z-score method
def detect_outliers_zscore(df, column, threshold=3):
    mean = df[column].mean()
    std = df[column].std()
    z_scores = (df[column] - mean) / std
    return df[np.abs(z_scores) > threshold]
zscore_outliers = {}
for column in df.select_dtypes(include=[np.number]).columns:
    zscore_outliers[column] = detect_outliers_zscore(df, column).shape[0]
print("Z-Score Outliers Count:")
print(zscore_outliers)
```

```python
# Remove outliers using IQR method
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
for column in df.select_dtypes(include=[np.number]).columns:
    data = remove_outliers_iqr(df, column)
iqr_cleaned_file = "dataset_cleaned_iqr.csv"
data.to_csv(iqr_cleaned_file, index=False)
print(f"Dataset cleaned using IQR saved at {iqr_cleaned_file}")
```
